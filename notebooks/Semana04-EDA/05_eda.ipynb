1. Cargar el Gold Dataset para el EDA

```python
gold_dataset = spark.read.format("delta").load(
    "/Volumes/workspace/default/churn_b2b/Gold/gold_dataset"
)
```

2. Revisar el esquema

```python
gold_dataset.printSchema()
```

3. Estadísticas descriptivas

```python
display(gold_dataset.describe())
```

4. Valores nulos

```python
from pyspark.sql.functions import col, count, when

gold_dataset.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in gold_dataset.columns
]).show()
```

5. Distribución del churn

```python
display(
    gold_dataset.groupBy("churn").count()
)
```

6. Empresas por sector

```python
display(
    gold_dataset.groupBy("sector").count()
)
```

7. Empresas por región

```python
display(
    gold_dataset.groupBy("region").count()
)
```

8. Estadísticas de facturación

```python
display(
    gold_dataset.select(
        "facturacion_total",
        "deuda_total",
        "trafico_total_gb"
    )
)
```

9. Churn por sector

```python
display(
    gold_dataset.groupBy("sector", "churn").count()
)
```

10. Churn por segmento

```python
display(
    gold_dataset.groupBy("segmento", "churn").count()
)
```

11. Correlaciones

```python
gold_dataset.stat.corr("facturacion_total", "deuda_total")
gold_dataset.stat.corr("cantidad_tickets", "churn")
gold_dataset.stat.corr("risk_score", "churn")
```
